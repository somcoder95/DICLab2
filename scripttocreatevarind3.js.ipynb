```python
import pandas as pd
df1 = pd.read_csv("wordcount.csv")
```

```python
df1
df1.iterrows()
```

```python
f= open("examples.words.js","w+")
f.write("var words =[")
for index,row in df1.iterrows():
    if (index!=(len(df1)-1)):
        f.write("{text: '%s', size: %s}," %(row['word'],row['count']))
    else :
        f.write("{text: '%s', size: %s} ];" %(row['word'],row['count']))
    
f.close()
```
